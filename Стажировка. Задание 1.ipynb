```python
import pandas as pd
import numpy as np
import matplotlib.pyplot as plt
import seaborn as sns
%matplotlib inline
df = pd.read_csv('C:\\Users\\egorka31\\Desktop\data_task3.csv', sep='\t')
```

Сначала реализуем обычную точность ответов

```python
tpr = df.groupby('uid')
q = [pd.DataFrame(tpr.get_group(x)) for x in tpr.groups]
accuracy = {}
for i in range(len(q)):
    x = q[i]
    accuracy[x['uid'].iloc[0]] =  len(x[x['jud'] == x['cjud']])/len(x)
```

Выпишем точность ассесора и его uid от худшего к лучшего

```python
for i in range(len(q)):
    print(sorted(accuracy.values())[i], sorted(accuracy, key=accuracy.get)[i])
```

Теперь реализуем точность с согласованностью

```python
df['true'] = [1 if df['jud'].iloc[i] == df['cjud'].iloc[i] else 0 for i in range(len(df))]
```

```python
docs = df.groupby('docid').sum()
```

```python
docs_count = df.groupby('docid').count()
```

```python
weight = {}
for i in range(len(docs)):
    weight[docs.index[i]] = docs['true'].iloc[i] / docs_count['true'].iloc[i]
```

```python
tpr = df.groupby('uid')
q = [pd.DataFrame(tpr.get_group(x)) for x in tpr.groups]
weight_accuracy = {}
for i in range(len(q)):
    x = q[i]
    good = 0
    all1 = 0
    for j in range(len(x)):
        if x['true'].iloc[j] == 1:
            good+= 1.2 - weight[x['docid'].iloc[j]]
            all1+= 1.2 - weight[x['docid'].iloc[j]]
        else:
            all1+= 0.2 + weight[x['docid'].iloc[j]]
    weight_accuracy[x['uid'].iloc[0]] =  good/all1
```

Выпишем точность (с учетом весов) ассесора и его uid от худшего к лучшего

```python
for i in range(len(q)):
    print(sorted(weight_accuracy.values())[i], sorted(weight_accuracy, key=weight_accuracy.get)[i])
```
